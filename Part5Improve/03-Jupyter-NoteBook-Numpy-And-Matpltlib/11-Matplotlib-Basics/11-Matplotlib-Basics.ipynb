# Matplotlib 绘图基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 把[0, 10]等分成100份
x
```

```python
y = np.sin(x) # 对x中的每一个值求sin
y
```

```python
plt.plot(x, y) # 以x为横坐标 
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape # cosy是含有100个元素的向量
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show() # 默认会进行颜色区分地
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red") # 也可以人为指定颜色
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle="--") # 也可以人为指定颜色，还可以人为指定样式
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle="--") # 也可以人为指定颜色，还可以人为指定样式
plt.xlim(-5, 15) # 可以人为指定x轴
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle="--") # 也可以人为指定颜色，还可以人为指定样式
plt.xlim(0, 1.5) # 可以人为指定y轴
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle="--") # 也可以人为指定颜色，还可以人为指定样式
plt.axis([-1, 11, -2, 2]) # 可以人为指定x、y轴范围，前两个是x轴范围，后两个是y轴范围
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle="--") # 也可以人为指定颜色，还可以人为指定样式
plt.xlabel("x axis") # 也可以给x、y轴添加说明
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x, siny, label = "sin(x)") # 可以给曲线加上含义
plt.plot(x, cosy, color = "red", linestyle="--", label = "cos(x)")
plt.xlabel("x axis") 
plt.ylabel("y axis")
plt.legend() # 声明这句才能给曲线加上label
plt.show()
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, color = "red", linestyle="--", label = "cos(x)")
plt.xlabel("x axis") 
plt.ylabel("y axis")
plt.legend()
plt.title("Hello My Name is LiangShanGuang")
plt.show()
```

# 绘制散点图(Scatter Plot)

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
x = np.random.normal(0, 1, 100)  # 生成均值为0方差为1(也称正太分布)的100个数
y = np.random.normal(0, 1, 100)  # 生成均值为0方差为1(也称正太分布)的100个数

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 100)  # 生成均值为0方差为1(也称正太分布)的100个数
y = np.random.normal(0, 1, 100)  # 生成均值为0方差为1(也称正太分布)的100个数

plt.scatter(x, y, alpha=0.4) # 可以人为指定透明度
plt.show()
```
